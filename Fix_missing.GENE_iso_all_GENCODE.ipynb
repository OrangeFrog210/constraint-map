```python
import pandas as pd
```

```python
f_iso_all = "iso_all_v1.1.csv"
df_iso_all = pd.read_csv(f_iso_all)
df_iso_all
```

## Steparate

```python
# df_iso_all[~df_iso_all["CHESS ID"].str.contains("hypothetical")]
```

```python
df_iso_all_no_GENE = df_iso_all[df_iso_all["GENE"] == "-"]
df_iso_all_no_GENE
```

```python
df_iso_all_yes_GENE = df_iso_all[df_iso_all["GENE"] != "-"]
df_iso_all_yes_GENE
```

```python
# Check if now rows are missed in the separation
print((df_iso_all_no_GENE.shape[0]+df_iso_all_yes_GENE.shape[0]) == df_iso_all.shape[0])
```

```python
# Write to csv file
f_no = "iso_all_v1.1_no_GENE.csv"
df_iso_all_no_GENE.to_csv(f_no, index=False)
```

```python
# Write to csv file
f_yes = "iso_all_v1.1_yes_GENE.csv"
df_iso_all_yes_GENE.to_csv(f_yes, index=False)
```

## Merge results back

```python
f_no_fixed = "iso_all_v1.1_no.GENE_filled.csv"
df_iso_all_no_GENE_fixed = pd.read_csv(f_no_fixed)
print(df_iso_all_no_GENE_fixed.shape[0] == df_iso_all_no_GENE.shape[0])
df_iso_all_no_GENE_fixed
```

```python
df_iso_all_no_GENE_fixed.columns = df_iso_all_no_GENE.columns
df_iso_all_no_GENE_fixed
```

```python
# Replace NaN with -
df_iso_all_no_GENE_fixed["GENE"] = df_iso_all_no_GENE_fixed["GENE"].fillna("-")
df_iso_all_no_GENE_fixed
```

```python
# Merge the df which originally were not missing GENE and the fixed formally-missing df
df_result = pd.concat([df_iso_all_yes_GENE, df_iso_all_no_GENE_fixed])
print(df_result.shape[0] == df_iso_all.shape[0])
df_result
```

```python
df_result = df_result.reset_index(drop=True)
df_result
```
