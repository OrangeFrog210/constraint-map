```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
f_base_gplmdca = "gplmDCA_protein_MSAsize_seqlen_bin_all.txt"
f_base_pydca = "pydca_protein_MSAsize_seqlen_bin_all.txt"
f_pydca = "output_pydca_all.txt"
f_gplmdca = "output_gplmDCA_all.txt"
```

```python
df_base_gplmdca = pd.read_csv(f_base_gplmdca, names=["Filename", "MSA_size", "Sequence_length", "Bin"])
df_base_pydca = pd.read_csv(f_base_pydca, names=["Filename", "MSA_size", "Sequence_length", "Bin"])
df_gplmdca = pd.read_csv(f_gplmdca, names=["Filename", "Execution_time", "Bin"])
df_pydca = pd.read_csv(f_pydca, names=["Filename", "Execution_time", "Bin"])
```

### pydca

```python
df_base_pydca
```

```python
df_base_pydca["Filename"] = df_base_pydca["Filename"].str.split('_').str[1].str.split('.').str[0]
df_base_pydca
```

## pydca

```python
df_pydca
```

```python
df_pydca["Filename"] = df_pydca["Filename"].str.split('_').str[1].str.split('.').str[0]
df_pydca
```

```python
print(df_pydca["Filename"].isna().sum())
df_pydca = df_pydca.dropna()
df_pydca
```

```python
df_pydca["h"] = df_pydca["Execution_time"].str.split('h').str[0]
df_pydca["m"] = df_pydca["Execution_time"].str.split('h').str[1].str.split('min').str[0]
df_pydca["s"] = df_pydca["Execution_time"].str.split('h').str[1].str.split('min').str[1].str.split('s').str[0]
df_pydca
```

```python
df_pydca["h"] = df_pydca["h"].astype(int)
df_pydca["m"] = df_pydca["m"].astype(int)
df_pydca["s"] = df_pydca["s"].astype(int)
#print(df_pydca.dtypes)
#print(df_pydca)
```

```python
df_pydca["Total_s"] = df_pydca["h"] * 3600 + df_pydca["m"] * 60 + df_pydca["s"]
df_pydca
```

## Merging each of the 2 dfs

## pydca

```python
# Merge pydca dfs
df_master_pydca = df_pydca.merge(df_base_pydca, how='inner', on='Filename')
df_master_pydca
```

```python
df_master_pydca = df_master_pydca[df_master_pydca["Sequence_length"] != 0]
```

### Re-binning based on seuqnece length

```python
# Re-grouping based on sequence length
df_master_pydca.loc[df_master_pydca['Sequence_length'] <= 99, 'seqlen_bin'] = '_99'
df_master_pydca.loc[df_master_pydca['Sequence_length'].between(100,199), 'seqlen_bin'] = '100_199'
df_master_pydca.loc[df_master_pydca['Sequence_length'].between(200,299), 'seqlen_bin'] = '200_299'
df_master_pydca.loc[df_master_pydca['Sequence_length'].between(300,499), 'seqlen_bin'] = '300_499'
df_master_pydca.loc[df_master_pydca['Sequence_length'].between(500,699), 'seqlen_bin'] = '500_699'
df_master_pydca.loc[df_master_pydca['Sequence_length'].between(700,899), 'seqlen_bin'] = '700_899'
df_master_pydca.loc[df_master_pydca['Sequence_length'].between(900,1099), 'seqlen_bin'] = '900_1099'
df_master_pydca.loc[df_master_pydca['Sequence_length'].between(1100,1299), 'seqlen_bin'] = '1100_1299'
df_master_pydca.loc[df_master_pydca['Sequence_length'].between(1300,1399), 'seqlen_bin'] = '1300_1399'
```

```python
df_master_pydca
```

```python
df_master_pydca.describe()
```

```python
binsize = 100
list_bins = [i for i in range (0, 1200, binsize)]
plt.hist(df_master_pydca["Sequence_length"], bins=list_bins)
plt.suptitle("Sequence length Distribution: n = " + str(df_master_pydca["Sequence_length"].shape[0]))
plt.title("bin size: " + str(binsize))
plt.xlabel("Sequence length")
plt.ylabel("Frequency")
```

```python
## Obtaining the results
df_abin = df_master_pydca[df_master_pydca["seqlen_bin"] == "1100_1299"]
df_abin
bins = [-1, 250, 500, 750, 1000, 1250, np.inf]
labels=["-250", "250-500", "500-750", "750-1000", "1000-1250", "1250-"]
df_abin_grouped = df_abin.groupby(pd.cut(df_abin["MSA_size"], bins=bins, labels=labels)).mean()
#df_abin_grouped = df_abin.groupby(pd.cut(df_abin["MSA_size"], bins=bins, labels=labels)).size().reset_index(name='count')
df_abin_grouped
```
