```python
df_iso_MANE_cim_filtered = df_iso_MANE_cim_filtered.sort_values(by="gene")
df_iso_MANE_cim_filtered
# df_iso_MANE_cim_filtered[(df_iso_MANE_cim_filtered["length_ratio"] == 1) & (df_iso_MANE_cim_filtered["gene"] == "-")]
```

```python
result = df_iso_MANE_cim_filtered.groupby('gene').agg({'aa_length_isoform': ['mean', 'min', 'max']}).reset_index()
  
print("Mean, min, and max values of aa length of transcript grouped by gene")
print(result)
```

```python
# rename columns
print(result.columns)
result.columns = ["gene", "mean aa length", "min aa length", "max aa length"]
print(result.columns)
result
```
