```python
import pandas as pd
import numpy as np
```

```python
# read iso_all
f_iso_all = "iso_all_v1.1_GENE.filled.csv"
df_iso_all = pd.read_csv(f_iso_all)
df_iso_all.columns = df_iso_all.columns.str.replace('CHESS ID', 'CHESS_ID')
df_iso_all
```

```python
# read iso_MANE
f_iso_MANE = "iso_MANE_cim_fixed.csv" # could be "iso_MANE_cim_fixed_nohypo.csv" for our purpose
df_iso_MANE = pd.read_csv(f_iso_MANE)
df_iso_MANE
```

```python
# Merge iso_all and iso_MANE
df_result = df_iso_all.merge(df_iso_MANE, how="inner", on="CHESS_ID")
df_result
```

```python
df_result["MANE_ALT"] = np.where(df_result["CHESS_ID"] == df_result["CHESS_ID_MANE"], "MANE", "ALT")
df_result
```

Since inner join resulted in same no. from MANE (fixed), it means all samples in MANE (fixed) are in iso_all

### Filter the merged df
#### 1. Keep only rows where there is transcript info in at least one of RefSeq and GENCODE

```python
# Show rows where RefSeq ID exist (not "-")
df_result[df_result["RefSeq ID"] != "-"].sample(10)
```

```python
# Show rows where GENECODE ID exist (not "-")
df_result[df_result["GENCODE ID"] != "-"].sample(10)
```

```python
# Count the number of rows where both RefSeq and GENCODE ID are "-"
df_result[((df_result["RefSeq ID"] == "-") & (df_result["GENCODE ID"] == "-")) & (df_result["MANE_ALT"] == "MANE")].shape
```

```python
df_result_filter1 = df_result[((df_result["RefSeq ID"] != "-") | (df_result["GENCODE ID"] != "-"))]
df_result_filter1
```

```python
# Check whether there are any rows containing "-" in both "RefSeq ID" and "GENCODE ID" cols
df_result_filter1[(df_result_filter1["RefSeq ID"] == "-") & (df_result_filter1["GENCODE ID"] == "-")]
```

```python
print((df_result[(df_result["RefSeq ID"] == "-") & (df_result["GENCODE ID"] == "-")].shape[0]
       +df_result_filter1.shape[0]) == df_result.shape[0] )
```

```python
print("The # of transcripts after filtering by REFSEX GENCODE IDs criterion is: ", str(df_result_filter1.shape[0]))
```

#### 2. Remove all hypothetical

```python
df_result_filter1[df_result_filter1["CHESS_ID"].str.contains("hypothetical")].shape
```

```python
df_result_filter1[(df_result_filter1["CHESS_ID"].str.contains("hypothetical")) & (df_result_filter1["MANE_ALT"] == "MANE") ] .shape
```

```python
df_result_filter2 = df_result_filter1[~df_result_filter1["CHESS_ID"].str.contains("hypothetical")]
df_result_filter2
```

```python
print((df_result_filter1[df_result_filter1["CHESS_ID"].str.contains("hypothetical")].shape[0]
       +df_result_filter2.shape[0]) == df_result_filter1.shape[0])
```

```python
# Confirming that the df is properly filtered
print(df_result_filter2[(df_result_filter2["RefSeq ID"] == "-") & (df_result_filter2["GENCODE ID"] == "-")])
print(df_result_filter2[df_result_filter2["CHESS_ID"].str.contains("hypothetical")])
```

```python
df_result_filter2 = df_result_filter2.drop_duplicates(subset=["CHESS_ID"])
```

```python
df_result_filter2["CHESS_ID"].to_csv("shortlisted_transcripts.txt", index=None, header=None)
```

```python
df_result_filter2.to_csv("merged.csv", index=None)
```
