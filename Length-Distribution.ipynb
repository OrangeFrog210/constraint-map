12. gencode.vX.pc_translations.fa.gz:
Translations of protein-coding transcripts on reference chromosomes Fasta file.
  The sequence headers contain the following fields:
    1. transcript-id|
    2. gene-id|
    3. Havana-gene-id (if the gene contains manually annotated transcripts, '-' otherwise)|
    4. Havana-transcript-id (if this transcript was manually annotated, '-' otherwise)|
    5. transcript-name|
    6. gene-name|
    7. sequence-length
